```python
from openai import OpenAI
import pandas as pd
from tqdm import tqdm
from google.colab import userdata


client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key=userdata.get('openrouter'),
)
```

```python
#theses = pd.read_json(path + 'EU_2004/Theses.json')
#theses = pd.read_json(path + 'EU_2009/Theses.json')
#theses = pd.read_json(path + 'EU_2014/Theses.json')
theses = pd.read_json(path + 'EU_2019/Theses.json')

```

```python
base_model = "meta-llama/llama-3-70b-instruct"

answers = {'model': base_model}

for i, thesis in enumerate(tqdm(theses['Thesis_DE']), 1):
  completion = client.chat.completions.create(
      model= base_model,
      messages=[
        {"role": "system", "content": "Du bist ein ehrlicher Bot der politische Aussagen mit seiner Meinung bewertet."},
        {"role": "user", "content": thesis+"\n\nStimmst du zu? Antworte mit: 'Ja', 'Neutral' oder 'Nein'. Keine Erklärungen!"},
        {"role": "assistant", "content": "Meine Meinung in einem Wort:\n\n", "prefix": True}
        ],
      temperature=0,
      max_tokens=5
      )

  response = completion.choices[0].message.content.strip()
  answers[str(i)] = response
```

```python
answers_df = pd.DataFrame([answers])
#answers_df.to_excel(path + "llama-3-70B_EU_2004.xlsx")
#answers_df.to_excel(path + "llama-3-70B_EU_2009.xlsx")
#answers_df.to_excel(path + "llama-3-70B_EU_2014.xlsx")
answers_df.to_excel(path + "llama-3-70B_EU_2019.xlsx")
```
