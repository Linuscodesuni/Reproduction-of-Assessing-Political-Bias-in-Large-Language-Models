### Alignment plots

```python
alignments_de = pd.read_excel(path + "alignment_de.xlsx")
```

```python



data = {
    'Political Party': alignments_de.columns[1:],
    'Llama2-7B':            alignments_de.loc[5][1:].apply(lambda x: round(x, 1)),
    'Llama3-8B':            alignments_de.loc[6][1:].apply(lambda x: round(x, 1)),
    'Llama3-70B':           alignments_de.loc[8][1:].apply(lambda x: round(x, 1)),
    'Mistral7B':            alignments_de.loc[7][1:].apply(lambda x: round(x, 1)),
    'Mistral7B_fix':         alignments_de.loc[9][1:].apply(lambda x: round(x, 1))
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert from wide to long format
df_long = pd.melt(df, id_vars=['Political Party'], var_name='LLM Model', value_name='Alignment')

# Define the custom color palette
party_colors = {
    "CDU": "#44448aff",
    "SPD": "#cc0033ff",
    "GRNE": "#32a460ff",
    "DIE LINKE": "#8f57c7ff",
    "AfD": "#36cac5ff",
    "CSU": "#0089c4ff",
    "FDP": "#ffcc33ff",
    "FREIE WHLER": "#8db4e3ff",
    "PIRATEN": "#ff6600ff",
    "Tierschutzpartei": "#92b64eff",
    "FAMILIE": "#dd1d7dff",
    "DP": "#ed9f0dff",
    "Die PARTEI": "#c74933ff",
    "Volt": "#7f1f5fff",
    "PDS": "#8f57c7ff"
}

# Set font size to be twice as big
font_size = 20

# Create the catplot
sns.set(style="whitegrid")
g = sns.catplot(
    data=df_long,
    x='LLM Model',
    y='Alignment',
    hue='Political Party',
    kind='bar',
    height=6,
    aspect=2,
    palette=party_colors
)

# Customize the plot with increased font size
g.set_axis_labels("LLM Model", "Alignment (%)", fontsize=font_size)
g.set_titles("LLM Models Alignment with Political Parties", fontsize=font_size)
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.xlabel("LLM Model", fontsize=font_size)
plt.ylabel("Alignment (%)", fontsize=font_size)

plt.savefig(path + 'Plots/alignments_german.svg')
```

```python
answers_excel_de = pd.read_excel(path + "models_de_coded.xlsx", index_col = "model")
answers_excel_en = pd.read_excel(path + "models_en_coded.xlsx", index_col = "model")
```

```python
answers_excel_de
```

```python
description_en_list = [item['Description_DE'] for item in json.load(open(path + 'theses.json'))]
description_en_list = [f'{i+1}. {description_en_list[i]}' for i in range(len(description_en_list))]
```

```python
answers_38B_de = answers_excel_de.loc[["meta-llama/Meta-Llama-3-8B-Instruct", "meta-llama/Meta-Llama-3-8B-Instruct_reproduction"]]

plt.figure(figsize=(12, 12))


# Define the custom colormap
colors = [ '#27b376', '#bf212f','#f9a73e']
cmap = sns.color_palette(colors)

# Create the heatmap
ax = sns.heatmap(answers_38B_de, annot=False, cbar=False,xticklabels=description_en_list,cmap=cmap,square=True,linewidths=0.5, linecolor='black')

ax.set_xticklabels(description_en_list, rotation=45, ha='right', rotation_mode='anchor')

# Add titles and labels
plt.xlabel('Questions')
plt.ylabel('Models')

plt.tight_layout()

# Show the plot
plt.savefig(path + 'Plots/answers_38B_de.svg')
```
