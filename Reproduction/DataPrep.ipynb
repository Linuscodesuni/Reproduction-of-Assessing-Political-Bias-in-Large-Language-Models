```python
df_statement = pd.DataFrame(statement)

Thesis_DE = list(df_statement["text"])
Description_DE = list(df_statement["label"])
ID = list(df_statement["id"])
```

```python
df_statement["id"] = df_statement["id"] + 1 # id der Statements soll bei 1 anfangen
```

```python
Theses = []

for x in enumerate(Thesis_DE):
    temp = {}
    temp["Thesis_DE"] = x[1]
    temp["Description_DE"] = Description_DE[x[0]]
    temp["ID"] = ID[x[0]]
    Theses.append(temp)
```

```python
with open(path + "EU_2004/Theses.json", "w", encoding = "utf-8") as f:
    json.dump(Theses,
             f,
             ensure_ascii = False)
```

```python

partyMap = {item['id']: item['name'] for item in party}


df_opinion = pd.DataFrame(opinion).pivot(index='party', columns='statement', values='answer')
df_opinion = df_opinion.rename(index = partyMap).copy()
```

```python

rename_dict = {c: int(int(c) + 1) for c in df_opinion.columns if str(c).isdigit()} #statements sollen bei id 1 anfangen
df_opinion.rename(columns=rename_dict, inplace=True)


parties = [
    "CDU", # verschiedene Optionen für CDU und CSU, da Sie verschieden gelistet werden
    "CSU",
    "CDU / CSU",
    "SPD",
    "GRÜNE",
    "Bündnis 90/Die Grünen",
    "DIE LINKE",
    "AfD",
    "FDP",
    "FREIE WÄHLER",
    "FW FREIE WÄHLER",
    "PIRATEN",
    "Tierschutzpartei",
    "Die Tierschutzpartei",
    "FAMILIE",
    "ÖDP",
    "ödp",
    "Die PARTEI",
    "Volt",
    "PDS"
]
df_opinion = df_opinion.loc[df_opinion.index.intersection(parties)]

```

```python

df_opinion = df_opinion.rename(index={"Bündnis 90/Die Grünen": "GRÜNE"})
df_opinion.to_excel(path + "party_opinions_2004.xlsx")
```
