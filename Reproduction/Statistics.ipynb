```python
score_1 = alignment_de.loc["meta-llama/Llama-2-7b-chat-hf_reproduction"].values
score_2 = alignment_de.loc["meta-llama/Llama-2-70b-chat-hf"].values
score_3 = alignment_de.loc["meta-llama/Meta-Llama-3-8B-Instruct_reproduction"].values
score_4 = alignment_de.loc["meta-llama/Meta-Llama-3-70B-Instruct_reproduction"].values
score_5 = alignment_de.loc["mistralai/Mistral-7B-Instruct-v0.2_reproduction"].values

kruskal_reproduction_de = kruskal(score_1, score_3, score_4, score_5)
```

```python
score_1 = alignment_de.loc["meta-llama/Llama-2-7b-chat-hf_reproduction"].values
score_2 = alignment_de.loc["meta-llama/Llama-2-70b-chat-hf"].values
score_3 = alignment_de.loc["meta-llama/Meta-Llama-3-8B-Instruct_reproduction"].values
score_4 = alignment_de.loc["meta-llama/Meta-Llama-3-70B-Instruct_reproduction"].values
score_5 = alignment_de.loc["mistralai/Mistral-7B-Instruct-v0.2_fixed"].values

kruskal_reproduction_fix = kruskal(score_1, score_3, score_4, score_5)

data = [score_1, score_3, score_4, score_5]
model_names = ["Llama 2-7b", "Llama 3-8b", "Llama 3-70b", "Mistral 7b"]

dunn_df = sp.posthoc_dunn(data, p_adjust='bonferroni')

dunn_df.columns = model_names
dunn_df.index = model_names

dunn_df
```
